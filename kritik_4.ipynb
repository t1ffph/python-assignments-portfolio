```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, num_iterations):#added number of iterations to better control code
                                                                    #if too many iterations jupyter will break lol or
                                                                    #control convergence speed and
                                                                    #if doesn't work within the number of iterations
                                                                    #i know to adjust the learning rate or other parameters
    
    def deriv(f, base_point): #estimate derivative at base point/current point of estimation
        return (f(base_point + 10**(-5)) - f(base_point)) / (10**(-5))
    
    x_coords = [initial_point] #storing x_n's
    y_coords = [f(initial_point)] #storing y_n's
    
    for _ in range(num_iterations):
        gradient = deriv(f, x_coords[-1]) #moving to the left bc if f'>0 then it increases to the right, looking for min
                                            #if start from left will move right
        update = -learning_rate * gradient #learning rate is some small #, recall h = −ηf′ and will continuously update
        new_x = x_coords[-1] + update #since update is already neg; and recall x1 = x0 − ηf′(x0). -1 for last/
                                    #most recent element in the list
        x_coords.append(new_x) #adding to list of x coords
        y_coords.append(f(new_x)) #adding to list of y coords

        
    #plotting info below, taken from greg + rp1 math labs
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000) #combined the lines greg gave us
    function_range = [f(i) for i in plot_range] #above is min and max in list
    
    plt.plot(plot_range, function_range, label='Function') #plotting f(x), just added a label for clarity
    plt.scatter(x_coords, y_coords, label='Gradient Descent') #plotting point sequences which is the descent
    plt.xlabel('x') #labelling the axes, below as well
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()
    
    return round(x_coords[-1], 3), round(y_coords[-1], 3) #returned variables rounded to 3 decimal places

def f(x): #this is where the function is inputted every time
    return abs(x)

#setting the parameters, which may be adjusted to better suit certain functions (see instructions)
learning_rate = 0.1
initial_point = 1
num_iterations = 2000

#perform the descent display the result that it finds
min_x, min_y = gradient_descent(f, learning_rate, initial_point, num_iterations)
print(f"x at predicted minimum: {min_x}, y at predicted minimum: {min_y}") #printing result!
```
