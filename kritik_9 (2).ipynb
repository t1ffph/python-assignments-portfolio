```python
import sympy as sp

#function 1
x, y = sp.symbols('x y')
f=sp.exp(x)*sp.sin(y)*y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("First partial WRT x:", df_dx)
print("First partial WRT y:", df_dy)
```

```python
import sympy as sp

#gradient vector
x, y = sp.symbols('x y')
g = x**2*y+x*y**2
dg_dx=sp.diff(g, x)
dg_dy=sp.diff(g, y)

x_value = 1
y_value = -1

dg_dx_value = dg_dx.subs({x: x_value, y: y_value})
dg_dy_value = dg_dy.subs({x: x_value, y: y_value})

print("First partial WRT x:", dg_dx_value)
print("First partial WRT y:", dg_dy_value)
print(" ")
print(f"Gradient vector: ({dg_dx_value}, {dg_dy_value})")
```

```python
import sympy as sp

#second partials
x, y = sp.symbols('x y')
h = sp.ln(x**2+y**2)
dh_dx=sp.diff(h, x)
dh_dy=sp.diff(h, y)

print("First partial WRT x:", dh_dx)
print("First partial WRT y:", dh_dy)

dh_dxx=sp.diff(dh_dx, x)
dh_dyy=sp.diff(dh_dy, y)
dh_dxy=sp.diff(dh_dx, y)
dh_dyx=sp.diff(dh_dy, x)

print(" ")
print("Second partial xx:", dh_dxx)
print("Second partial yy:", dh_dyy)
print("Second partial xy:", dh_dxy)
print("Second partial yx:", dh_dxy)
print(" ")
print("according to Clariut's theorem for mixed partials, as long as the second derivatives are continuous for the neighbourhood being investigated, the mixed partials for the function will be the same. If there are discontinuoties, the mixed partials will not be symmetrical.")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#contour plot example
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

#contour plot of another function
x, y = sp.symbols('x y')
t = sp.ln(x**2)+2*y*x

t_func = sp.lambdify((x, y), t, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = t_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $t(x,y) = ln(x^2)+2xy$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
